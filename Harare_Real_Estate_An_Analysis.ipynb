```python
"""
source files 
"""

# imf 
yearly_zimbabwe_inflation_file = 'data/imf/imf-inflation-rate-1980-2025.xls'
yearly_zimbabwe_population = 'data/imf/imf-population-1980-2025.csv'

# world bank
personal_remmitances_zimbabwe = 'data/world-bank/zim-remmitances-1977-2024.csv'


# zimstat 
housing_and_population_harare_2012 = 'data/zimstat/housing_population_data_2012.csv'
housing_and_population_harare_2022 = 'data/zimstat/housing_population_data_2022.csv'

# real estate 
average_property_prices_harare = 'data/property/property_prices.csv'
```

```python
"""
import relevant modules and load files 
"""

import pandas as pd 
import matplotlib.pyplot as plt
import csv


yearly_inflation_df = pd.read_excel(yearly_zimbabwe_inflation_file)
yearly_inflation_df.drop([0, 2, 3], inplace=True)
years = pd.Series(yearly_inflation_df.columns.to_list()[1:46])

inflation = yearly_inflation_df.iloc[0].to_list()[1:46]

yearly_inflation_df = pd.DataFrame(data={'Year': years, 'Inflation': inflation})
yearly_inflation_df['Year'] = pd.to_numeric(yearly_inflation_df['Year'], errors='coerce')
yearly_inflation_df['Inflation'] = pd.to_numeric(yearly_inflation_df['Inflation'], errors='coerce')


yearly_inflation_df.plot.scatter(x='Year', y='Inflation')
```

```python
# population 

yearly_zimbabwe_population_df = pd.read_csv(yearly_zimbabwe_population)
yearly_zimbabwe_population_df = pd.DataFrame(data={'Year': yearly_zimbabwe_population_df.columns.to_list()[7:52], 'population': yearly_zimbabwe_population_df.iloc[0].to_list()[7:52]})
yearly_zimbabwe_population_df['Year'] = pd.to_numeric(yearly_zimbabwe_population_df['Year'], errors='coerce')
yearly_zimbabwe_population_df['population'] = pd.to_numeric(yearly_zimbabwe_population_df['population'], errors='coerce')
yearly_zimbabwe_population_df.plot.scatter(x='Year', y='population')
```

```python
# remmitances 

# having problems reading using pandas so will use work around 
#personal_remmitances_zimbabwe_df = pd.read_csv(personal_remmitances_zimbabwe, on_bad_lines='skip')
#personal_remmitances_zimbabwe_df

with open(personal_remmitances_zimbabwe, 'r') as f:
    reader = csv.reader(f)
    remmitances_data = list(reader)
    years = remmitances_data[4][25:-1]
    zim_remmitances = remmitances_data[-1][25:-1]

personal_remmitances_zimbabwe_df = pd.DataFrame(data={'Year': years, 'remmitances': zim_remmitances})

personal_remmitances_zimbabwe_df['Year'] = pd.to_numeric(personal_remmitances_zimbabwe_df['Year'], errors='coerce')
personal_remmitances_zimbabwe_df['remmitances'] = pd.to_numeric(personal_remmitances_zimbabwe_df['remmitances'], errors='coerce')

personal_remmitances_zimbabwe_df.plot.scatter(x='Year', y='remmitances')
```
